```python
import os
import numpy as np
import tensorflow as tf
from tensorflow.keras.preprocessing import image
from tensorflow.keras.applications.resnet50 import ResNet50, preprocess_input
from sklearn.metrics.pairwise import cosine_similarity
from google.colab import files
import matplotlib.pyplot as plt
from PIL import Image

def load_and_preprocess_image(image_path):
    img = image.load_img(image_path, target_size=(224, 224))
    img_array = image.img_to_array(img)
    img_array = np.expand_dims(img_array, axis=0)
    return preprocess_input(img_array)

def extract_image_embedding(model, img_path):
    img = load_and_preprocess_image(img_path)
    return model.predict(img)

def search_similar_images(query_image_path, dataset_path):

    base_model = ResNet50(weights='imagenet', include_top=False, pooling='avg')


    query_embedding = extract_image_embedding(base_model, query_image_path)


    similar_images = []
    for filename in os.listdir(dataset_path):
        if filename.endswith(('.jpg', '.jpeg', '.png')):
            image_path = os.path.join(dataset_path, filename)
            dataset_embedding = extract_image_embedding(base_model, image_path)
            similarity_score = cosine_similarity(query_embedding, dataset_embedding)[0][0]


            if similarity_score > 0.8:
                similar_images.append((filename, similarity_score))


    similar_images.sort(key=lambda x: x[1], reverse=True)


    query_img = Image.open(query_image_path)
    plt.figure(figsize=(12, 6))
    plt.subplot(2, 3, 1)
    plt.imshow(query_img)
    plt.title('Query Image')
    plt.axis('off')


    for i in range(min(5, len(similar_images))):
        img_path = os.path.join(dataset_path, similar_images[i][0])
        img = Image.open(img_path)
        plt.subplot(2, 3, i + 2)
        plt.imshow(img)
        plt.title(f"Similar Image {i + 1}\n(Similarity Score: {similar_images[i][1]:.4f})")
        plt.axis('off')

    plt.tight_layout()
    plt.show()


uploaded = files.upload()


query_image_filename = list(uploaded.keys())[0]


search_similar_images(query_image_filename, '/content/drive/MyDrive/Colab Notebooks/fossil dataset')
```
